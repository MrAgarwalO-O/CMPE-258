```python
import pandas as pd

# Synthetic data for a fictional football league
data = {
    "Player Name": ["Elon Musk", "Jeff Bezos", "Mark Zuckerberg", "Sundar Pichai", "Satya Nadella"],
    "Team": ["SpaceX Rockets", "Amazon Warriors", "Facebook FC", "Google Strikers", "Microsoft United"],
    "Nationality": ["South Africa", "USA", "USA", "India", "India"],
    "Goals": [36, 28, 22, 26, 30],
    "Assists": [12, 14, 19, 17, 15],
    "Passes Completed": [1900, 2000, 1800, 2100, 1950],
    "Matches Played": [34, 34, 34, 34, 34],
    "Yellow Cards": [3, 2, 4, 1, 2],
    "Red Cards": [0, 1, 0, 0, 0]
}


df = pd.DataFrame(data)
df
```

```python
df = df.astype(str)
```

```python
from transformers import AutoTokenizer, AutoModelForTableQuestionAnswering, pipeline
model = AutoModelForTableQuestionAnswering.from_pretrained(
                "google/tapas-large-finetuned-wtq")
```

```python
tokenizer = AutoTokenizer.from_pretrained("google/tapas-large-finetuned-wtq")
nlp = pipeline('table-question-answering', model=model, tokenizer=tokenizer)
```

```python
question_list = [
    "Who scored the highest number of goals?",
    "How many assists were made by Sundar Pichai?",
    "Which player has the least yellow cards?",
    "What is the total number of red cards received by players from Amazon Warriors?",
    "Who has the highest passes completed, and how many passes did they complete?",
    "How many goals were scored by players from India?",
    "What is the average number of matches played by all players?",
    "Who received a red card from Microsoft United?",
    "Which team does the player with the highest number of assists play for?",
    "What is the total number of goals scored by all players?"
]
```

```python
result = nlp({'table': df, 'query': question_list[0]})
print(result)
```

```python
for question in question_list:
    result = nlp({'table': df, 'query': question})
    print(result['cells'][0].strip())
```
